```python
df=df.drop(['size','area_type'],axis='columns')
```

```python
def avg_sqft(x):
    token=x.split('-')
    if len(token)==2:
        return (float(token[0])+float(token[1]))/2
    try:
        return float(x)
    except:
        return None
    
```

```python
df.total_sqft=df.total_sqft.apply(avg_sqft)
```

```python
df['price_per_sf']=(df.price*100000/df.total_sqft)
```
